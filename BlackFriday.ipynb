```python
#imports

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_train = '/Users/simrangadkari/Downloads/train.csv'
path_test = '/Users/simrangadkari/Downloads/test.csv'
```

```python
train_df = pd.read_csv(path_train)
test_df = pd.read_csv(path_test)

train_df.head(10)
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
train_df.info()
```

```python
train_df.describe()
```

```python
#Fill the NaNs in ProductCategory2 and ProductCategory3 with the mean value of respective columns


mean_value1=train_df['Product_Category_2'].mean()
mean_value2=train_df['Product_Category_3'].mean()
# print(mean_value1)

train_df['Product_Category_2'].fillna(value=mean_value1, inplace=True)

train_df['Product_Category_3'].fillna(value=mean_value2, inplace=True)


train_df.head(10)



```

```python
train_df.drop(['User_ID','Product_ID'], axis=1, inplace=True)

#Removing the '+' from Age and Stay_In_Current_City_Years columns

train_df['Age']=(train_df['Age'].str.strip('+'))

train_df['Stay_In_Current_City_Years']=(train_df['Stay_In_Current_City_Years'].str.strip('+').astype('float'))


train_df.head()
```

```python
#Using Label Encoder for encoding Categorical data in Gender, Age and City_Category as numerical data 

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

train_df.Gender = encoder.fit_transform(train_df.Gender)
train_df.Age = encoder.fit_transform(train_df.Age)
train_df.City_Category = encoder.fit_transform(train_df.City_Category)

# train_df[:,0] = encoder.fit_transform(train_df[:,0])

train_df.head(25)
```

```python
#Train Valid split using the Training data csv

from sklearn.model_selection import train_test_split

X = train_df.iloc[:,0:9].values
y = train_df.iloc[: , -1].values

# print(X)
# print(y)

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.1, random_state = 0)

# print(X_train.shape)
# print(X_valid.shape)
# print(y_train.shape)
# print(y_valid.shape)
```

```python
#Standardizing the data in all columns to fit to one scale

from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

X_train = std_scaler.fit_transform(X_train)

X_valid = std_scaler.fit_transform(X_valid)

# print(X_valid)
```

```python
#model used - Linear Regression 
#Error metric used - R2 Squared Error

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

reg = LinearRegression()
  
reg.fit(X_train, y_train)

yhat = reg.predict(X_valid)

# print(yhat)
# print(y_valid)
r2 = r2_score(y_valid, yhat)

print(r2)

```

```python
#Adjusted R2 score
print(1 - (1-reg.score(X, y))*(len(y)-1)/(len(y)-X.shape[1]-1))
```
